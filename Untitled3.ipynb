# Text preprocessing

```python
cleaned_sentences=pd.Series(sentences).str.replace("[^a-zA-Z]", " ")
```

```python
cleaned_sentences = [s.lower()for s in cleaned_sentences]
```

```python
stop_words=stopwords.words('english')
```

```python
stop_words
```

```python
def remove_stopwords(cleaned_sentences):
    new_s=" ".join([i for i in cleaned_sentences if i not in stop_words])
    return new_s
```

```python
cleaned_sentences =[remove_stopwords(i.split()) for i in cleaned_sentences]
```

```python
cleaned_sentences
```

# use Glove pre_trained word vector 

```python
word_embedding={}
file=open('E:\/glove.6B.100d.txt' , encoding='utf-8')
```

```python
for line in file:
    values =line.split()
    word=values[0]
    coefs=np.asarray(values[1:] , dtype='float32')
    word_embedding[word]=coefs
file.close()
```

```python
sentence_vectors = []
for i in cleaned_sentences:
    if len(i) != 0:
        vector = sum([word_embedding.get(w, np.zeros((100,))) for w in i.split()]) / len(i.split())+ 0.001
    else:
        vector = np.zeros((100,))
    sentence_vectors.append(vector)
print(sentence_vectors)
```

```python
sim_mat=np.zeros([len(sentences),len(sentences)])
```

```python
#generate similarity matrix
for i in range(len(sentences)):
    for j in range(len(sentences)):
        if i == j:
            sim_mat[i][j] = cosine_similarity(sentence_vectors[i].reshape(1,100), sentence_vectors[j].reshape(1,100))[0,0]
```

```python
nx_data=nx.from_numpy_array(sim_mat)
scores=nx.pagerank(nx_data)
```

```python
ranked_sentences=sorted(((scores[i] , s)for i , s in  enumerate(sentences)),reverse=True)
```
